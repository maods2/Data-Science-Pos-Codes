## Consultando um banco de dados usando acesso nativo

```python
# Importando o conector MYSQL para Python
import mysql.connector as msql
```

```python
# Estabelecendo uma conexão com banco de dados
conexao_nativa = msql.connect(user='root', password='labsqlbd', database='sakila')
```

```python
# Consultando o banco de dados
cursor = conexao_nativa.cursor()
cursor.execute('SELECT title, release_year FROM film ORDER BY title')
```

```python
# Verificando alguns atributos do cursor
print(f'Total de registros: {cursor.rowcount}')
print(f'Tamanho padrao do bloco de registro: {cursor.arraysize}')
```

```python
# Buscando um registro
regs = cursor.fetchone()
if regs:
    print(regs)
```

```python
# Buscando varios registros um a um
regs = cursor.fetchmany(5)
if regs:
    print(regs)
```

```python
# Buscando todos os registros - CUIDADO! - :)
regs = cursor.fetchall()
if regs:
    print(regs)
```

```python
# Atribuindo os registros a um Data Frame do Pandas
import pandas as pd
cursor.execute('select first_name, email from customer order by first_name')
regs = cursor.fetchall()
df = pd.DataFrame.from_records(regs, columns=['filme', 'lancamento'])
df.head(10)
```

```python
cursor.close()
conexao_nativa.close()
```
