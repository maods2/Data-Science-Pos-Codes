### 09. Calcule o total de vendas de cada produto por semestre e apresente-os usando dois gráficos de pizza. Cada gráfico deve trazer o rótulo e o valor percentual de cada fatia.



O gráfico deve ser semelhante a esse:

image.png

```
primeiro_semestre = df.iloc[:6,1:-2].sum()
primeiro_semestre_total = df.iloc[:6,-2].sum()

segundo_semestre = df.iloc[6:,1:-2].sum()
segundo_semestre_total = df.iloc[6:,-2].sum()

pie_chart_1sem = pd.DataFrame(round(primeiro_semestre / primeiro_semestre_total * 100,2)).T
pie_chart_2sem = pd.DataFrame(round(segundo_semestre / segundo_semestre_total * 100,2)).T

fig, ax = plt.subplots(1, 2)

ax[0].set_title('Vendas no 1º Semestre')
ax[0].pie(pie_chart_1sem.iloc[0].values, labels=pie_chart_1sem.columns, autopct='%1.1f%%', startangle=0)
ax[0].axis('equal')


ax[1].set_title('Vendas no 2º Semestre')
ax[1].pie(pie_chart_2sem.iloc[0].values, labels=pie_chart_2sem.columns, autopct='%1.1f%%', startangle=0)
ax[1].axis('equal')
plt.show();
```

#### Existe algum diferença destacável na proporção de vendas entre os semestres? Responda com um print()

```
'''
Alguns produtos como o Sabonete, pasta de dente e Mascara Facial apresentaram uma variação significativa na proporção de vendas entre o primeiro e segundo semestre,
o que indica que esses itens estão sujeitos a influência sazonal quando observamos suas respectivas demandas.

'''

fig, ax = plt.subplots(1, 3,figsize=(30,8))
ax[0].bar(df.month_number, df['bathingsoap'], label='Vendas de sabonete', color ='g', alpha=0.4)
ax[1].bar(df.month_number, df['toothpaste'], label='Vendas de Pasta de dente', alpha=0.4)
ax[2].bar(df.month_number, df['facecream'], label='Vendas de Mascara Facial',color ='r', alpha=0.4 )
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.show()
```

### 10. Exiba o total de vendas de cada produto ao longo dos meses em um gráfico de área. Defina um grau de transparência ao gráfico para deixar a visualização mais agradável.

O gráfico deve ser semelhante a esse:

image.png

```
colors = ['royalblue','orange','seagreen','indianred','mediumpurple','sienna']

x = df['month_number']

y1 = df.iloc[:, 1]
y2 = df.iloc[:, 2]
y3 = df.iloc[:, 3]
y4 = df.iloc[:, 4]
y5 = df.iloc[:, 5]
y6 = df.iloc[:, 6]

fig, ax = plt.subplots()
ax.stackplot(x,y1,y2,y3,y4,y5,y6, labels=df.iloc[:,1:-2].columns, colors=colors, alpha=0.7)
ax.legend(loc='upper left')
plt.show()
```
