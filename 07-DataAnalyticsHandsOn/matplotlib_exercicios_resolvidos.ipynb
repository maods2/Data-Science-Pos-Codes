### 09. Calcule o total de vendas de cada produto por semestre e apresente-os usando dois gráficos de pizza. Cada gráfico deve trazer o rótulo e o valor percentual de cada fatia.



O gráfico deve ser semelhante a esse:

image.png

```
labels = ['Creme facial', 'Sabonete de rosto', 'Pasta de dente', 'Sabonete', 'Shampoo', 'Hidratante']
sales_semestre_1 = [sales['facecream'].iloc[:6].sum(), sales['facewash'].iloc[:6].sum(),
                    sales['toothpaste'].iloc[:6].sum(), sales['bathingsoap'].iloc[:6].sum(),
                    sales['shampoo'].iloc[:6].sum(), sales['moisturizer'].iloc[:6].sum()]

sales_semestre_2 = [sales['facecream'].iloc[6:].sum(), sales['facewash'].iloc[6:].sum(),
                    sales['toothpaste'].iloc[6:].sum(), sales['bathingsoap'].iloc[6:].sum(),
                    sales['shampoo'].iloc[6:].sum(), sales['moisturizer'].iloc[6:].sum()]

fig, (ax1, ax2) = plt.subplots(1,2)
ax1.pie(sales_semestre_1, labels=labels, autopct='%1.1f%%')
ax2.pie(sales_semestre_2, labels=labels, autopct='%1.1f%%')
fig.subplots_adjust(hspace=0.4, wspace=0.4)
fig.set_size_inches(15,10)
ax1.set_title('Vendas no 1º semestre')
ax2.set_title('Vendas no 2º semestre')
plt.show()
```

### 10. Exiba o total de vendas de cada produto ao longo dos meses em um gráfico de área. Defina um grau de transparência ao gráfico para deixar a visualização mais agradável.

O gráfico deve ser semelhante a esse:

```
labels = ['Creme facial', 'Sabonete de rosto', 'Pasta de dente', 'Sabonete', 'Shampoo', 'Hidratante']
plt.stackplot(sales.month_number, sales.facecream, sales.facewash, sales.toothpaste,
              sales.bathingsoap, sales.shampoo, sales.moisturizer, labels=labels, alpha=0.7)
plt.legend(loc='upper left')
plt.show()
```

### 11. Compare o volume de vendas de hidratante e creme facial com um gráfico de barras.

O gráfico de barras deve exibir o número de unidades vendidas por mês para cada produto. Adicione uma barra separada para cada produto no mesmo gráfico.

Neste mesmo gráfico, apresente o lucro total de cada mês. Como a escala de quantidade de produtos é diferente da escala de lucro, crie um eixo gêmeo (pesquise por twin axes).




O gráico deve ser semelhante a esse:

image.png

```
fig, ax1 = plt.subplots()

ax1.bar([a-0.4 for a in sales.month_number], sales.moisturizer, width= 0.4, label = 'Vendas de Hidratante', align='edge')
ax1.bar([a for a in sales.month_number], sales.facecream, width= 0.4, label = 'Vendas de Creme facial', align='edge')
ax1.grid(False)
ax_twin = ax1.twinx()


ax_twin.plot(sales.month_number, sales.total_profit,
             marker='o', color='g', markersize=10, linewidth=4,
             markerfacecolor='w')

ax1.set_xlabel('Mês')
ax1.set_xticks(sales.month_number)
ax1.set_ylabel('Unidades vendidas')
ax1.legend(loc='upper left')
ax1.set_title('Vendas de hidratante e creme facial');
```

### 12. Crie uma grade de subplots e, para cada painel, exiba um gráfico de dispersão relacionando os dados de vendas de cada produto e o total de lucro obtido no mês. Na última linha da grade deve ser apresentado um único painel com um gráfico de caixas exibindo a distribuição de vendas de cada produto.

```
produtos = {'facecream':'Creme facial', 'facewash':'Sabonete de rosto',
          'toothpaste':'Pasta de dente', 'bathingsoap':'Sabonete',
          'shampoo':'Shampoo', 'moisturizer':'Hidratante'}
```

```
colunas = list(produtos.keys())
```

```
a = sales.loc[:, colunas]
a
```

```

fig = plt.figure(figsize=(15, 20))
grid = plt.GridSpec(4, 3, wspace=0.4, hspace=0.3)

for i in range(2):
    for j in range(3):
      ax = fig.add_subplot(grid[i, j])
      ax.scatter('total_profit', colunas[(i+j)], data=sales)
      ax.set_xlabel('Lucro total obtido')
      ax.set_ylabel('Vendas de ' + produtos[colunas[(i+j)]])


ax = fig.add_subplot(grid[2:, :])
sales.loc[:, colunas].boxplot(ax=ax)
```
